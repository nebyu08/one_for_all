# lets see how to draw a tangent line 

```python
def f2(x):
    return 2*x**2

x=np.arange(0,5,0.001)
y=f2(x)

#lets calculate the derivative
delta=0.001
x1=2
x2=x1+delta

y1=f2(x1)
y2=f2(x2)

derivative=(y2-y1)/(x2-x1)

#y=mx+b
b=y2-derivative*x2
```

```python
def tangent_line(x):
    """the derivative and the y-intercept are the same only the  """
    return derivative*x+b
```

```python
plot_points=[x1-0.9,x1,x1+0.9]

plt.plot(x,y)
plt.plot(plot_points,[tangent_line(x) for x in plot_points])
plt.show()
```

```python
print(f"the apporiximate derivative for x is, {x1} and {derivative}")
```

# draw multiplie tangent lines

```python
# lets draw multiple tangent lines for the graph
x=np.arange(0,5,0.001)
y=f2(x)

colors=['k','g','r','b','c']

#lets plot the graph here
plt.plot(x,y)

#delta increment
delta=0.001

#drawing lines using slope and x points
def tangent_lines_v2(x,derivatives,b):     # .==> y=mx+b
    return derivatives*x+b

for i in range(5):
    #inputs
    x1=i
    x2=i+delta
    
    #outputs
    y1=f2(x1)
    y2=f2(x2)

    derivatives=(y2-y1)/(x2-x1)
    b=y1-derivatives*x1
    
    tobe_plotted=[x1-0.9,x1,x1+0.9]
    
    plt.scatter(x1,y1,c=colors[i])
    plt.plot([point for point in tobe_plotted],[tangent_lines_v2(point,derivatives,b) for point in tobe_plotted],c=colors[i])
    print(f"the input is {x1} and its derivative is {derivatives:} ")

plt.show()
```

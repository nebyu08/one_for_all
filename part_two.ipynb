# Multi-Layer neural networks

```python
inputs=[
    [1,2,3,2.5],
    [2,5,-1,2],
    [-1.5,2.7,3.3,-0.8]
]

weight_1=[
    [0.2,0.8,0.4,-1],
    [0.5,0.9,0.26,-0.5],
    [-0.26,-0.27,0.17,0.87]
]

# ==> 3x4 . 4x3 = 3x3

bias_1=[-1,2,0.5]


weight_2=[
    [0.1,0.14 ,0.5],
    [-0.5,0.12,-0.33],
    [-0.44,0.73,-0.13]
]

bias_2=[-1,2,-0.5]

#our multi-layer neural network

layer1_output=np.dot(inputs,np.array(weight_1).T)+bias_1

layer2_output=np.dot(layer1_output,np.array(weight_2).T)+bias_2
print(layer2_output)
```

# Dense layers

```python
class Dense:
    """
    this is the structure of a single hidden layer that is meant to be customized
    """
    def __init__(self,n_inputs,n_neurons):
        self.weights=0.01*np.random.randn(n_inputs,n_neurons)
        self.bias=np.zeros((1,n_neurons))
    def forward(self,inputs):
        """ this is the feedforward opertaion of a neural network."""
        self.outputs=np.dot(inputs,self.weights)+self.bias
        return self.outputs
```
